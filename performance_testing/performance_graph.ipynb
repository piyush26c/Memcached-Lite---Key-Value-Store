```python
import matplotlib.pyplot as plt

# Data for Memcached-lite Server (read and write latency)
memcache_data = {
    'Number of Key-Value Pairs': [1, 10, 100, 1000, 10000, 100000],
    'Read Latency': [0.00284910202, 0.002691745758, 0.006350755692, 0.03760671616, 0.3002314568, 4.294315338],
    'Write Latency': [0.002312898636, 0.001910924911, 0.007217407227, 0.05722570419, 0.8464083672, 6.702391148]
}

# Data for Google Key Value Store Server (read and write latency)
google_kv_data = {
    'Number of Key-Value Pairs': [1, 10, 100, 1000, 10000, 100000],
    'Read Latency': [0.2613654137, 0.2453660965, 0.3891282082, 0.5255310535, 1.592638254, 7.278608322],
    'Write Latency': [0.5271482468, 0.5519337654, 0.8615503311, 1.158763885, 2.609219551, 16.89333367]
}

# Corresponding file sizes for the number of key-value pairs
file_sizes = ['4KB', '46KB', '372KB', '4.1MB', '41MB', '410.7MB']

```

```python
# Create separate figures for each plot
plt.figure(figsize=(20, 10))
# Plot Read Latency for Memcached-lite Server (Exponential)
plt.semilogy(memcache_data['Number of Key-Value Pairs'], memcache_data['Read Latency'], marker='o', label='Memcached-lite Server')
plt.xlabel('Number of Key-Value Pairs')
plt.ylabel('Read Latency (s)')

# Annotate the points with read latency values
for i, (x, y) in enumerate(zip(memcache_data['Number of Key-Value Pairs'], memcache_data['Read Latency'])):
    offset = (0, 10)
    if i > 0:
        prev_x, prev_y = memcache_data['Number of Key-Value Pairs'][i - 1], memcache_data['Read Latency'][i - 1]
        if abs(x - prev_x) < 10:  # Adjust the threshold for overlapping points
            offset = (0, -20)  # Move the annotation below the point
    plt.annotate(f'({x}, {y:.6f} s)', (x, y), textcoords="offset points", xytext=offset, ha='center')


plt.title('Read Latency vs. Number of Key-Value Pairs (Memcached-lite Server)')
plt.grid(True)
plt.show()
```
